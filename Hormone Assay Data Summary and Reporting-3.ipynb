```python
import pandas as pd
import matplotlib.pyplot as plt


csvfile = "Desktop/hormone_assay_data.csv"

# For comma-separated
df = pd.read_csv(csvfile, delimiter=',')

df
```

## Handling Missing Values
Missing hormone values were first accounted for, then replaced with the column mean to preserve all samples for analysis.


```python
print("\nMissing values per column:")
print(df.isna().sum())
```

```python
#fix missing value
for col in ['FSH', 'LH', 'Testosterone']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df
```

```python
df=df.set_index('sample_id')
```

```python
df
```

```python
df.describe()
```

```python
summary=df.describe()
```

```python
#Get Testosterone_status
def testosterone_status(value):
    if value < 3.0:
        return 'Low'
    elif value <= 4.0:
        return 'Normal'
    else:
        return 'High'

df['Testosterone_status'] = df['Testosterone'].apply(testosterone_status)
```

```python
df
```

## CONCLUSION
Most samples fall within the normal testosterone range, suggesting typical hormonal balance in the dataset

```python
#Count samples in each category
status_counts = df['Testosterone_status'].value_counts()
print(status_counts)
```

```python
# Saved clean dataset and summary stats as CSV
df.to_csv("hormone_cleaned.csv", index=False)
summary.to_csv("hormone_summary_stats.csv")
```

```python
#  Plot hormone distributions
plt.figure(figsize=(12,6))

for i, col in enumerate(['FSH','LH','Testosterone'], 1):
    plt.subplot(1,3,i)
    plt.hist(df[col], bins=8, color='skyblue', edgecolor='black')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
```

```python
# Plot Testosterone_status counts
plt.figure(figsize=(6,4))
status_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Testosterone Status Counts')
plt.xlabel('Status')
plt.ylabel('Number of Samples')
plt.show()
```
